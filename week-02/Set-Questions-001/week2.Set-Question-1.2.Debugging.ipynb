```python
#9. Find the error in the following Python code:
# x = 10
# if x = 5:
#     print("x is 5")
#------------------------------------------

#
# Error:
#   1.SyntaxError: invalid syntax. Maybe you meant '==' or ':=' instead of '='?
#
# solution: 
#     change the assignment operator (=) into  equality comparison operator (==)

print("using equality comparison operator ==")
x = 10
if x == 5:
    print("x is 5")
else:
    print("x is",x)


#
# solution: 
#  change the assignment operator (=) into  assignment expression (:=)
#
print("using  assignment expression := ")
x = 10
if x := 5:
    print("x is 5")

#
# use the solution according to requirement
#
```
