```python
# 4.Scenario: In a secure login system, certain features are enabled only 
# for users with prime-numbered user IDs.
#  Write logic to check if a given number is prime.

num1 = int(input("enter a number to check if it is prime: "))
is_prime=True
for i in range(2,num1):
    if num1 == 1 or num1 == 2 or num1 == 3:
       break 
    if num1%i ==0:
      is_prime=False
      break
if is_prime:
    print(num1," is prime")
else:
    print(num1, " is not prime")
```

```python
# 5.Scenario: A scientist is working on permutations and needs 
# to calculate the factorial of numbers frequently.
#  Write logic to find the factorial of a given number using recursion.

def fact(n):
    return 1 if n <=1 else n * fact(n-1)

fact_number=int(input("please enter a integer: "))
fact_val=fact(fact_number)
print()
print("The Factorial to be calculated for: ",fact_number)
print("The Factorial of",fact_number,"is : ",fact_val)
```

```python
#  6.Scenario: A unique lottery system assigns ticket numbers 
#  where only Armstrong numbers win the jackpot.
#  Write logic to check whether a given number is an Armstrong number.
#---------------------------------------------------------------------------
# a number that is equal to the sum of its own digits 
# each raised to the power of the number of digits. 
# For example, 153 is an Armstrong number because: 1^3 + 5^3 + 3^3 = 153.
# 1, 2, 3, 4, 5, 6, 7, 8, 9, and 153, 370

str1 = input("enter a number to check if it is Armstrong number: ")
num_str1 = int(str1)
len_str1=int(len(str1))
list1=list(str1)
#print(list1)
total=0
for i in range(0,len(list1)):
    total += int(list1[i]) **len_str1

print("Number Inputted :",int(str1))
print("The Corresponding Armstrong compute sum: ", total)
if (int(str1) == total):
    print ("Number :", int(str1),"is an Armstrong number")
else:
    print ("Number :", int(str1),"is NOT an Armstrong number")
```

```python
# 7.Scenario: A password manager needs to strengthen weak 
# passwords by swapping the first and last characters of user-generated passwords.
#  Write logic to perform this operation on a given string.

str1 = input("enter a password that need to be strengthen: ")
list1=list(str1)
first_character=list1[0]
last_character=list1[-1]
#swap characters
#print(list1)
list1[0]=last_character
list1[-1]=first_character
#print(list1)
s_str1 = "".join(list1)
print()
print("-----------------------------")
print("Given Password : ",str1)
print("Strengthened Password: ",s_str1) 
print("-----------------------------")
```
