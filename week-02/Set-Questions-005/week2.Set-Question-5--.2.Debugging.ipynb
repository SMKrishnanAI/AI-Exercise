```python
#5.What is the issue with this function?

def is_prime(n):
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

num1=7
is_prime_match=""
if not is_prime(num1):
    is_prime_match= "[NOT]"
print(f"The number {num1} is {is_prime_match} a prime")

# when a number has a factor then it is not prime
# the return inside if statement should return False and return outside if should return True
```

```python
#6.What is the issue with this function?

def filter_even(numbers):
    return [num for num in numbers if num % 2 == 1]

def filter_odd(numbers):
    return [num for num in numbers if num % 2 == 0]

print(f"The even numbers are filtered from the list and produce odd numer list: {filter_even([1,2,3,4,5])}")
print(f"The odd numbers are filtered from the list and produce even numer list: {filter_odd([1,2,3,4,5])}")
#if filters the even number and sends out odd ones in the list
#if  if num % 2 == 0, it will filter out odd values and send the even numbers
```

```python
#9.What is the issue with this recursive function?

def factorial(n):
    #return n * factorial(n-1)
    return 1 if n <=1 else n * factorial(n-1)

num1=5
print(f"factorial of {num1} is {factorial(num1)}")

#RecursionError: maximum recursion depth exceeded
# function does not prescribe any return integer return value at any point in time
# so the recursion continues beyond required limit

#solution 
#include a if condition and return n when the recursion value is less than are equal to 1
```
