```python
# BMI Ranges for Adults
# BMI
# WEIGHT STATUS
# Below 18.5 Underweight
# 18.5-24.9  Healthy Weight
# 25.0-29.9  Overweight
# 30.0-34.9  Obese Class 1
# 35.0-39.9 Obese Class 2
# 40.0 and higher Obese Class 3
```

```python
bmi = float(input("Please Enter the BMI index:"))
if(bmi < 18.5):
    print("Underweight")
elif(bmi < 25):
    print("Healthy Weight")
elif(bmi < 30):
    print("Overweight")
elif(bmi < 35):
    print("Obese Class 1")
elif(bmi < 40):
    print("Obese Class 2")
else:
    print("Obese Class 3")
```
