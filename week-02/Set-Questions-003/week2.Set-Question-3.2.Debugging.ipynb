```python
#7.What will happen when you run the following code?

# numbers = [1, 2, 3, 4]
# numbers.sort().reverse()
# print(numbers)

numbers = [1, 2, 3, 4]
numbers.sort()
numbers.reverse()
print(numbers)



#In Python, the .sort() and .reverse() methods are in-place operations. 
#This means they modify the list directly and return None.
# so when numbers.sort() you have None and None does not have .


#solution
# # to make it work  
# numbers.sort().reverse() 
# can be broken down as
# numbers.sort()
# numbers.reverse()
```
